```python
def load_doc(id_):
    path = "mdenny_copy_early2015/bills/POS_Tagged_Bill_Versions/bill{}.anno".format(id_)
    print("Loading".format(id_))
    with open(path) as f:
        doc = json.load(f)
        sentences = [s['tokens'] for s in doc['sentences']]
        return pd.Series([sentences], index=[id_])
```

```python
def get_sentences(number_docs):
    name = 'load_doc'
    return dask.dataframe.Series(
        dsk={(name, id_): (load_doc, id_) for id_ in range(number_docs)},
        name=name,
        divisions=range(number_docs + 1),
        meta=pd.Series([[]])
    )
```
